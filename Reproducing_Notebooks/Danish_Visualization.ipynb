```python
fig = filein[filein.character_name == 'nora'].plot(x='Line_num', y='distribution_nora', color='orange', label='nora')
filein[filein.character_name == 'not-nora'].plot(x='Line_num', y='distribution_not-nora', color='blue', label='not-nora', ax=fig)

fig.set_xlabel('Line Number over Time')
fig.set_ylabel('Distribution Values')
fig.set_title('Distribution over Play Timeline')

fig=plt.gcf()
fig.set_size_inches(20, 8)
plt.show()
```
