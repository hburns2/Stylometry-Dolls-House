```python
import csv
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import nltk
```

```python
filein = pd.read_csv("Weka_Edited_Bokmal.csv")
filein.head()
```

```python
nora_data = filein[ filein["character_name"] == "nora " ]
nora_data.head()
```

```python
fig = nora_data[nora_data.character_name == 'nora'].plot(kind='scatter', x='line_num' y='distribution_nora', color='orange', label='Nora')

fig.set_xlabel('Nora Distribution')
fig.set_ylabel('Line Number')
fig.set_title('Noral Distribution vs Line Number')

fig=plt.gcf()
fig.set_size_inches(10, 7)
plt.show()
```
